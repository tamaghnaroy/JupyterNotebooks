```python
import FinAPI.MacroData.Employment as emp_api

datasets = emp_api.Data

JOLTS = emp_api.load_historical_data_dict(datasets.JOLTS)
```

```python
JOLTS[['vacancy_rate','hire_rate','quits_rate', 'layoff_rate', 'total_sep_rate']].corr()
```

```python
JOLTS= JOLTS.assign(vacancy_rate = JOLTS['Job Openings'] / (JOLTS['Job Openings'] + JOLTS['Total Employed']) * 100)
JOLTS = JOLTS.assign(hire_rate=JOLTS['Job Openings']/JOLTS['Total Employed'] * 100)
JOLTS = JOLTS.assign(quits_rate=JOLTS['Quits']/JOLTS['Total Employed'] * 100)
JOLTS = JOLTS.assign(layoff_rate=JOLTS['Layoffs']/JOLTS['Total Employed'] * 100)
JOLTS = JOLTS.assign(total_sep_rate =(JOLTS['Quits'] + JOLTS['Layoffs'] + JOLTS['Other Separation'])/JOLTS['Total Employed'] * 100)
JOLTS = JOLTS.assign(unemp_rate=JOLTS['Total Unemployed']/ (JOLTS['Total Employed'] + JOLTS['Total Unemployed']) * 100)
JOLTS = JOLTS.assign(date=JOLTS.index)
```

```python
alt.Chart(JOLTS).mark_circle().encode(alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative')
                                     ).properties(width=100, height=150).\
    repeat(row=['vacancy_rate','hire_rate','quits_rate', 'layoff_rate', 'total_sep_rate'],
            column=['vacancy_rate','hire_rate','quits_rate', 'layoff_rate', 'total_sep_rate']).interactive()
```
